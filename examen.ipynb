# Parte 1

```python
bikes=pd.read_excel('/content/drive/MyDrive/Data Scientist/examen practico /Technical_Assignment/Technical_Assignment/Bike.xlsx')
bike_shop=pd.read_excel('/content/drive/MyDrive/Data Scientist/examen practico /Technical_Assignment/Technical_Assignment/Bike_shop.xlsx')
order=pd.read_excel('/content/drive/MyDrive/Data Scientist/examen practico /Technical_Assignment/Technical_Assignment/orders.xlsx')
```

```python
bike_shop.keys()
```

```python
tienda=bike_shop.merge(order,left_on='bikeshop.id',right_on='order.id',how='inner')
tienda_tipos=tienda.merge(bikes,left_on='product.id',right_on='bike.id',how='inner')
tienda_tipos.head()
```

```python
tienda_tipos['category']=0
```

```python
for i in range(len(tienda_tipos)):
  if tienda_tipos['description'][i][0:8] == 'Mountain':
    tienda_tipos['category'][i]= 'Mountain'
  elif tienda_tipos['description'][i][0:4] == 'Road':
    tienda_tipos['category'][i]= 'Road'
  else:
    tienda_tipos['category'][i]= np.nan
tienda_tipos.head()
```

```python
tienda_tipos['Revenue']=tienda_tipos['quantity']*tienda_tipos['price']
tienda_tipos['Revenue'].head()
```

# Parte 2

```python
#columnas en el Dataframe
tienda_tipos.keys()
```

```python
#modelo supersix
supersix=tienda_tipos.loc[tienda_tipos['model'].str.contains('Supersix') == True]
supersix['States']=[supersix['location'].str.split(', ').iloc[i][1] for i in range(len(supersix))]
supersix.index=supersix['order.date']
supersix.head(8)
```

# Parte 3

```python
cuantiles=supersix['Revenue'].quantile([.25, .75])
```

```python
supersix['Revenue Quantiles']=0
for i in range(len(supersix)):
  if supersix['Revenue'].iloc[i]> cuantiles[0.75]:
    supersix['Revenue Quantiles'].iloc[i]='High'
  elif (supersix['Revenue'].iloc[i]<= cuantiles[0.75]) & (supersix['Revenue'].iloc[i]>= cuantiles[0.25]):
    supersix['Revenue Quantiles'].iloc[i]='Medium'
  elif supersix['Revenue'].iloc[i]< cuantiles[0.25]:
    supersix['Revenue Quantiles'].iloc[i]='Low'

supersix.head(10)
```
